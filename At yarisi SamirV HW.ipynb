```python
import random

class Horse_Rice:
    def __init__(self, ad, yasi, cinsi, rang, olke):
        self.ad = ad
        self.yasi = yasi
        self.cinsi = cinsi
        self.rang = rang
        self.olke = olke
        self.suret = 0
        self.mesafe = 0
    
    def suret_artir(self):
        if self.suret < 60:
            self.suret += random.randint(3, 10)
        elif self.suret >= 60:
            self.suret -= 5
        self.mesafe += self.suret 
        
    def suret_azalt(self):
        if self.suret > 0:
            self.suret -= 5
    
    def qet_edilen(self):
        return self.mesafe

    def status(self):
        print(f"Ad: {self.ad}")
        print(f"Yasi: {self.yasi}")
        print(f"Cinsi: {self.cinsi}")
        print(f"Rang: {self.rang}")
        print(f"Olke: {self.olke}")
        print(f"Suret: {self.suret}")
        print(f"Qet edilen mesafe: {self.mesafe}")

atlar = [
    Horse_Rice("At 1", 5, "Qarabag", "Qara", "Azerbaycan"),
    Horse_Rice("At 2", 6, "Bred", "Qahvayi", "ABS"),
    Horse_Rice("At 3", 4, "Iraq", "Ag", "Arab")
]

print('Zahmat olmasa atinizi secin!!')
for i, at in enumerate(atlar, 1):
    print(f"{i}. {at.ad}")
while True:
    try:
        secim = int(input('Atin nomresini qeyd edin: '))
        if 1 <= secim <= len(atlar):
            secilen_at = atlar[secim - 1]
            break
        else:
            print("Sehv secdiniz!")
    except ValueError:
        print("Zehmet olmasa duzgun nomre daxil edin!")

qalib = None
while not qalib:
    for horse in atlar:
        horse.suret_artir()
        horse.status()
        print("-" * 30)
        if horse.qet_edilen() >= 100:
            qalib = horse
            break

for horse in atlar:
    if horse == qalib:
        print(f"{horse.ad} Qalib geldin!\nTabrik edirem!!!")
    else:
        print(f"{horse.ad} Uduzdun!")
```
