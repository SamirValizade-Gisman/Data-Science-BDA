```python
data['duration'] = data['duration'].str.replace(" min","")
gr_data = data.groupby(['duration'],as_index=False)[['duration']]\
                .mean().rename(columns={"duration":"duration_avg"})
```

```python
gr_data
```

```python
data['duration'] = data['duration'].astype(float)
merged_data = pd.merge(data, gr_data, left_on='duration', right_on='duration_avg', how='left')
merged_data.head()
```
